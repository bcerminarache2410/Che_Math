# Problem 2 Linearization

You're given the nonlinear, first-order, time-invariant ODE $\frac{dy}{dt} = sin(y)$.

1. In the space below, plot the flow diagram for the ODE. 

```python
import math
```

```python
y = np.linspace(-100,100,1000)
dydt = np.zeros(len(y))

for i in range(len(y)):
    dydt[i] = math.sin(y[i])
```

```python
plt.plot(y,dydt,'r');
plt.ylabel('dy/dt');
plt.xlabel('y');
plt.axhline(y=0);
plt.title('dydt as a Function of y');
plt.legend(['dydt','y=0']);
plt.xlim([-10,10]);
plt.ylim([-2,2]);
```

2. In the space below, type out the equation for the linearized form of our ODE when it is linearized about the lowest, positive-valued, stable steadystate.

#### The lowest, positive value, stable steady state can be found at $y~=3.14$

#### Therefore, $y' = y - 3.14$ and $dy'/dt = sin(y')$, which has a steady state at the origin!

#### So, the linearized form of our ode is: $dy'/dt = sin(y-3.14)*(y-3.14)$

3. We want to compare the responses of our linearized model to our original nonlinear model. On a single figure, plot the time response of both models when starting at an initial conditions of $y_0 = 3$. Clearly indicate which lines represent which ODEs.

```python
timesteps = np.linspace(0,10,100)
y0 = [3.,3.]
def dydt(y,t):
    myy1 = y[0]
    myyprime = y[1]
    
    dydt = math.sin(myy1)
    dyprimedt = math.sin(myyprime-3.14)*(myyprime-3.14)
    return [dydt,dyprimedt]
```

```python
y = odeint(dydt,y0,timesteps)
```

```python
plt.plot(y);
plt.ylabel('y');
plt.xlabel('Time');
# plt.axhline(y=0);
plt.title('y as a Function of Time');
plt.legend(['Non-linearized','Linearized']);
# plt.xlim([-10,10]);
# plt.ylim([-2,2]);
```
